```python
#libraries
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
import numpy as np
```

```python
#LOAD DATASET
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
#Normalizing pixels values to a range of [0,1]
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0
```

```python
# Building the Neural NW model

model = Sequential()
```

```python
# transform the 28x28 images and flatten them, into vectors of size 784
model.add(Flatten(input_shape=(28,28)))

# define the hidden layer with 128 neurons, along with the Relu activation function
model.add(Dense(128, activation='relu'))

# define the output layer with 10 neurons and the softmax activation function
model.add(Dense(10, activation='softmax'))
```

```python
# Compilation and training of the model

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']) # the metric of performance is the accuracy

history = model.fit(
    x_train,
    y_train,
    epochs=5,    # if the accuracy is wrong, we can increase to improve it
    batch_size=32,
    validation_split = 0.1  # only use the 10% of the training data of the dataset for validation
)
```

```python
# Evaluating the model and testing
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)
print(f"Text accuracy: {test_acc:.4f}")


# testing the model by using one example to predict an image
index = np.random.randint(0, len(x_test)) # any index from 0 to the lenght of the test
sample = x_test[index:index+1] # for keeping dimensions
prediction = model.predict(sample)


print("REAL LABEL: ", y_test[index])
print("PREDICTED LABEL", np.argmax(prediction))
```
